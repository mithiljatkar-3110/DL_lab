GaussianNB

```python
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
url = "https://raw.githubusercontent.com/krishnaik06/Naive-Bayes/master/iris.csv"
```

```python
from sklearn.datasets import load_iris

# try to load from URL, fallback to sklearn's built-in iris if the URL fails
try:
    df = pd.read_csv(url)
    # ensure the target column is named 'species' for later cells
    if 'species' not in df.columns and 'target' in df.columns:
        df = df.rename(columns={'target':'species'})
except Exception:
    iris = load_iris()
    df = pd.DataFrame(iris.data, columns=iris.feature_names)
    df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
df.tail()
```

```python
X = df.drop('species',axis = 1)
```

```python
y = df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
nb = GaussianNB()
```

```python
nb.fit(X_train,y_train)
```

```python
y_pred = nb.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

BernaulliNB

```python
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
```

```python
breast_cancer = load_breast_cancer()
```

```python
X, y = breast_cancer.data, breast_cancer.target
```

```python
X
```

```python
y
```

```python
X_scaled = StandardScaler().fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y,test_size=0.2, random_state=42)
```

```python
nb = BernoulliNB(binarize=0.0)
```

```python
nb.fit(X_train, y_train)
```

```python
y_pred = nb.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
